```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
```

```python
df.info()
```

# 1. Scatter Plot (Numerical vs Numerical)

```python
sns.scatterplot(x='Age',y='Fare',data=df)
```

```python
sns.scatterplot(x='Age',y='Fare',data=df,hue='Sex')
```

```python
sns.scatterplot(x='Age',y='Fare',data=df,hue='Sex',style='Pclass')
```

```python
sns.scatterplot(x='Age',y='Fare',data=df,hue='Sex',style='Pclass',size='Survived')
```

# 2. Bar Plot (Numerical- Categorical)


Pclass me average age kitne h we use bar plot

```python
df.head()
```

```python
sns.barplot(x='Pclass',y='Age',data=df)
```

gender ke hisab se average fair kitna h har class ka

```python
sns.barplot(x='Pclass',y='Fare',hue='Sex',data=df)
```

# 3. Box Plot(Numerical -Categorical)

```python
sns.boxplot(x='Sex',y='Age',data=df,hue='Survived')
```

# 4. Displot (Numerical- Categorical)

```python
sns.displot(x='Age',data=df,kde=True)
```

```python
sns.displot(df[df['Survived']==0]['Age'],kde=True,)
sns.displot(df[df['Survived']==1]['Age'],kde=True)
```

# 5. Heat Map(Categorical-Categorical)

```python
pd.crosstab(df['Pclass'],df['Survived'])
```

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

```python
(df.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

# 6. Cluster map(Categorical-Categorical)

```python
pd.crosstab(df['Parch'],df['Survived'])
```

```python
sns.clustermap(pd.crosstab(df['Parch'],df['Survived']))
```

# 7. Pair Plot

Pair plot data me se total jitne bhi numerical columns hote h unme scatter plot plot kr deta h automatically

```python
sns.pairplot(df)
```
